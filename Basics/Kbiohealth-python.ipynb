형 변환(casting)이란, 한 데이터 타입을 다른 데이터 타입으로 변환하는 것입니다.

* 사용하는 함수: int(), float(), str(), list(), set() 등
* 예를 들어, 문자열로 된 숫자는 연산을 하기 전에 정수(int)나 실수(float) 로 변환해야 한다.
* 숫자와 문자열을 합치고 싶다면, 먼저 숫자를 문자열(str) 로 변환해야 한다.

```python
# str -> float, int
num_str = '10.6'
num_float = float(num_str)
print('num_float:', num_float)       # 10.6

num_int = int(num_float)
print('num_int:', num_int)           # 10
```

```python
# int -> str
num_int = 10
num_str = str(num_int)
print(num_str)                       # '10'
```

```python
# int -> float
num_int = 10
print('num_int:', num_int)           # 10
num_float = float(num_int)
print('num_float:', num_float)       # 10.0

# float -> int
gravity = 9.81
print(int(gravity))                  # 9
```

## 02-2 연산 (기본형, 배열형)

|Operator|Name|Example|
|:--:|:--:|:--:|
|+|Addition|x + y|
|-|Subtraction|x - y|
|*|Multiplication|x * y|
|/|Division|x / y|
|%|나머지 연산자(Modulus)|x % y|
|**|거듭제곱(Exponentiation)|x ** y|
|//|몫 연산자(Floor division)|x //y |

```python
# 숫자형의 연산(1/3): 정수

print('Addition: ', 1 + 2)
print('Subtraction: ', 2 - 1)
print('Multiplication: ', 2 * 3)

print("")

print('Division: ', 4 / 2)        # 파이썬에서는 나누기 연산이 항상 실수(float)를 반환
print('Division: ', 6 / 2)
print('Division: ', 7 / 2)

print("")

print('Division without the remainder: ', 7 // 2)   # 몫만 반환
print('Division without the remainder: ', 7 // 3)
print('Modulus: ', 3 % 2)         # 나머지
print('Exponentiation: ', 2 ** 3) # 8 -> 2 * 2 * 2
```

```python
# 숫자형의 연산(2/3): 실수
print('Floating Point Number, PI', 3.14)
print('Floating Point Number, gravity', 9.81)
```

```python
# 숫자형의 연산(3/3): 복소수
print('Complex number: ', 1 + 1j)
print('Multiplying complex numbers: ', (1 + 1j) * (1 - 1j))
```

**Question: 리스트와 튜플은 무엇이 다른가요?**

* 리스트와 튜플은 서로 다른 자료형을 담을 수 있는 **순서가 있는** 컬렉션입니다.
* 다만, 튜플은 **불변(immutable)** 이므로, 한 번 생성하면 값을 변경할 수 없습니다.
* 따라서 add, insert, remove 같은 메서드는 사용할 수 없습니다.

-> 튜플은 불변이므로 수정하려면 리스트로 변환 후 수정 가능 <br>
-> 튜플의 개별 항목은 삭제 불가. 전체는 삭제 가능


```python
# 튜플 예시 (1): 염기서열 위치 정보 (Genomic coordinates)
# VCF(Variant Call Format)에서 변이 위치를 고유하게 식별하기 위해 튜플로 저장
# VCF 형식: (염색체, 위치, REF, ALT)

vcf_variants = [
    ("chr1", 100000, "A", "G"),      # SNP
    ("chr2", 200000, "AT", "A"),     # INDEL
    ("chr17", 43044295, "C", "T"),   # BRCA1 변이
]
print(vcf_variants)
```

```python
# 튜플 예시 (2): 게놈 영역 정보
# BED(Browser Extensible Data) 파일에서 유전자 위치를 정의할 때 튜플로 저장
# BED 형식: (염색체, 시작, 끝)

bed_regions = [
    ("chr1", 0, 100000),
    ("chr2", 100000, 200000),
    ("chr17", 43044295, 43125364),   # BRCA1 위치
]
print(bed_regions)
```

# 03 생물정보학에서의 데이터 타입

파이썬의 다양한 데이터 타입은 생물정보학 데이터 분석에 필수적입니다.
각 타입이 생물학적 데이터에서 어떻게 사용되는지 살펴봅시다.

```python
# ============================================
# 1. 문자열 (str) - 염기서열, 단백질 서열
# ============================================

# DNA 염기서열
dna_sequence = "ATGCGATCGATCG"
print(f"DNA 서열: {dna_sequence}")
print(f"서열 길이: {len(dna_sequence)} bp")

# 단백질 서열
protein_sequence = "MKVLWAALLVTFAGCAKSDQKQ"
print(f"단백질 서열: {protein_sequence}")
print(f"서열 길이: {len(protein_sequence)} aa")

# 유전자 이름
gene_name = "BRCA1"
print(f"유전자 이름: {gene_name}")
```

```python
# ============================================
# 2. 정수 (int) - 개수, 위치, 좌표
# ============================================

# 염기 개수 세기
dna = "ATGCGATCGATCG"
a_count = dna.count('A')
t_count = dna.count('T')
print(f"A 개수: {a_count}")
print(f"T 개수: {t_count}")

# 염색체 번호
chromosome = 17  # BRCA1은 17번 염색체
print(f"BRCA1 위치: 염색체 {chromosome}")

# 시작 위치, 끝 위치
start_position = 43044295
end_position = 43125364
print(f"유전자 길이: {end_position - start_position} bp")

# 환자 나이
patient_age = 45
print(f"환자 나이: {patient_age}세")
```

```python
# ============================================
# 3. 실수 (float) - 확률, 점수, 비율
# ============================================

# GC 함량 (%)
gc_content = 45.5
print(f"GC 함량: {gc_content}%")

# E-value (BLAST 검색 결과 신뢰도)
e_value = 2.3e-50
print(f"E-value: {e_value}")

# 진화 거리 (유사도)
identity = 98.7
print(f"염기 서열 상동성: {identity}%")

# 단백질 농도 (mg/ml)
protein_concentration = 2.5
print(f"단백질 농도: {protein_concentration} mg/ml")
```

```python
# ============================================
# 4. 불린 (bool) - 참/거짓
# ============================================

# 유전자 발현 여부
is_expressed = True
is_mutated = False
print(f"유전자 발현됨: {is_expressed}")
print(f"돌연변이 감지됨: {is_mutated}")

# 질병 위험도 판단
has_mutation = True
is_pathogenic = True

if has_mutation and is_pathogenic:
    print("⚠️ 병원성 변이 감지됨!")
else:
    print("정상")
```

```python
# ============================================
# 5. 리스트 (list) - 여러 서열, 샘플
# ============================================

# 여러 DNA 샘플
dna_samples = [
    "ATGCGATCGATCG",
    "ATGCGATCGATCG",
    "ATGCGATCGATCG"
]
print("DNA 샘플 3개:")
for i, sample in enumerate(dna_samples, 1):
    print(f"  샘플 {i}: {sample}")

# 질병 관련 단백질 목록
disease_proteins = ["TP53", "BRCA1", "BRCA2", "MTHFR"]
print(f"질병 관련 단백질: {disease_proteins}")

# 환자 나이 리스트 (임상 데이터)
patient_ages = [25, 45, 67, 38, 52]
average_age = sum(patient_ages) / len(patient_ages)
print(f"평균 나이: {average_age:.1f}세")
```

```python
# ============================================
# 6. 튜플 (tuple) - 불변 좌표, 고정 정보
# ============================================

# 유전자 위치 (변경하면 안 됨)
gene_location = ("chr17", 43044295, 43125364)
print(f"유전자 위치: {gene_location}")
print(f"  염색체: {gene_location[0]}")
print(f"  시작: {gene_location[1]}")
print(f"  끝: {gene_location[2]}")

# VCF 형식 (variant 정보 - 기본 정보는 고정)
variant_key = ("chr1", 100000, "A", "G")
print(f"변이 정보: {variant_key}")

# 염기쌍 (A-T는 항상 쌍)
base_pairs = ("A", "T")
print(f"염기쌍: {base_pairs}")
```

```python
# ============================================
# 7. 세트 (set) - 중복 제거, 교집합
# ============================================

# 환자 A가 가진 변이
patient_a_mutations = {"TP53", "BRCA1", "MTHFR", "TP53"}  # TP53 중복
print(f"환자 A 변이 (중복 제거): {patient_a_mutations}")

# 환자 B가 가진 변이
patient_b_mutations = {"BRCA1", "BRCA2", "MTHFR"}
print(f"환자 B 변이: {patient_b_mutations}")

# 공통 변이 (교집합)
common_mutations = patient_a_mutations & patient_b_mutations
print(f"공통 변이: {common_mutations}")

# 환자 A만 가진 변이 (차집합)
unique_to_a = patient_a_mutations - patient_b_mutations
print(f"환자 A만 가진 변이: {unique_to_a}")
```
